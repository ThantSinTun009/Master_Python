```python
# Simple generator

def gen_values():
    yield 1
    yield 2
    yield 3
    
gen_values() # -- object
```

```python
gen_values_obj = gen_values()
print(next(gen_values_obj))
print(next(gen_values_obj))
print(next(gen_values_obj))
```

```python
print(next(gen_values_obj))
```

```python
# use for loop

for value in gen_values():
    print(value)
```

### Find Prime numbers

**Eager Evaluation**

```python
def get_primes_list(start, end):
    primes = []
    for num in range(start, end+1):
        if num < 2:
            continue
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
        if is_prime:
                primes.append(num)
    return primes
    
get_primes_list(1, 7)
```

With yield...

**Lazy Evaluation**

```python
def get_primes_list(start, end):
    for num in range(start, end + 1):
        if num < 2:
            continue
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            yield num  

for i in get_primes_list(1, 7):
    print(i)
```

```python
get_primes_obj = get_primes_list(1, 7)
print(next(get_primes_obj))
print(next(get_primes_obj))
print(next(get_primes_obj))
print(next(get_primes_obj))
```

In Generators, value is one at a time. You can't reuse the values you have outputed.
