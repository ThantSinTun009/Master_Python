# List Comprehension



```python
num_list = [1, 3, 5, 7, 9]

sqr_list = [n ** 2 for n in num_list]

print(sqr_list)
```

```python
tv_show = ['firends', 'Park and Recreation', 'Happy life', 'avengers: age of ultron']

cap_tv_show = [text.title() for text in tv_show]
print(cap_tv_show)
```

# With conditionals

```python
cap_tv_show = [text.title() for text in tv_show if len(text) >= 15]

print(cap_tv_show)
```

# Create a list from scratch

```python
[i ** 3 for i in range(1, 11)]
```

## Performance benefits

```python
# for runtime
from time import time
```

```python
start = time()
[i ** 3 for i in range(1, 10000001)]
end=time()
print("List comp run time:", end - start)
```

```python
start = time()
squares=[]
for n in range(1, 10000001):
    squares.append(n**3)
end=time()
print("Loop + append run time:", end - start)
```
